```python
if start_coords and end_coords:
    html_code = f"""<!DOCTYPE html>
    <html>
    <head>
        <title>IIT Indore Navigation</title>
        <meta charset="utf-8" />
        <meta name="viewport" content="width=device-width, initial-scale=1.0">

        <!-- Leaflet CSS -->
        <link rel="stylesheet" href="https://unpkg.com/leaflet@1.2.0/dist/leaflet.css" />
        <link rel="stylesheet" href="https://unpkg.com/leaflet-routing-machine@latest/dist/leaflet-routing-machine.css" />

        <style>
            body {{ margin: 0; padding: 0; }}
            #map {{ height: 80vh; width: 100%; }}
            .legend {{
                margin-top: 10px;
                padding: 10px;
                background: white;
                border-radius: 5px;
                font-size: 14px;
                border: 1px solid #ccc;
            }}
            .legend div {{
                display: flex;
                align-items: center;
                margin-bottom: 5px;
            }}
            .legend span {{
                width: 12px;
                height: 12px;
                display: inline-block;
                margin-right: 5px;
                border-radius: 50%;
            }}
            .blue-marker {{ background: blue; }}
            .red-marker {{ background: red; }}
        </style>
    </head>
    <body>
        <div id="map"></div>

        <!-- Leaflet JS -->
        <script src="https://unpkg.com/leaflet@1.2.0/dist/leaflet.js"></script>
        <script src="https://unpkg.com/leaflet-routing-machine@latest/dist/leaflet-routing-machine.js"></script>

        <script>
            var map = L.map('map').setView([{start_coords[0]}, {start_coords[1]}], 15);

            L.tileLayer('https://{{s}}.tile.openstreetmap.org/{{z}}/{{x}}/{{y}}.png', {{
                attribution: '© OpenStreetMap contributors'
            }}).addTo(map);

            var start = L.latLng({start_coords[0]}, {start_coords[1]});
            var end = L.latLng({end_coords[0]}, {end_coords[1]});

            // Google Maps-style marker icons
            var blueIcon = L.icon({{
                iconUrl: 'https://maps.google.com/mapfiles/kml/paddle/blu-circle.png',
                iconSize: [40, 40],
                iconAnchor: [20, 40]
            }});

            var redIcon = L.icon({{
                iconUrl: 'https://maps.google.com/mapfiles/kml/paddle/red-circle.png',
                iconSize: [40, 40],
                iconAnchor: [20, 40]
            }});

            // Add start location marker (Blue)
            L.marker(start, {{ icon: blueIcon }}).addTo(map).bindPopup("Start Location");

            // Add end location marker (Red)
            L.marker(end, {{ icon: redIcon }}).addTo(map).bindPopup("End Location");

            // Routing control
            var control = L.Routing.control({{
                waypoints: [start, end],
                routeWhileDragging: true,
                createMarker: function() {{ return null; }}  // Disable default markers
            }}).addTo(map);

            control.on('routesfound', function(e) {{
                var container = document.querySelector('.leaflet-routing-container');
                if (container) {{
                    var legend = document.createElement('div');
                    legend.className = 'legend';
                    legend.innerHTML = `
                        <div><span class="blue-marker"></span> Start Location</div>
                        <div><span class="red-marker"></span> Destination</div>
                    `;
                    container.appendChild(legend);
                }}
            }});
        </script>
    </body>
    </html>"""

    # Save to file
    with open("index.html", "w") as f:
        f.write(html_code)

    # Function to start the local HTTP server
    def start_server():
        os.system("python -m http.server 8001")

    # Start the server in a background thread
    server_thread = threading.Thread(target=start_server, daemon=True)
    server_thread.start()

    # Wait for a moment to let the server start
    time.sleep(2)

    # Open the webpage automatically in the default browser
    webbrowser.open("http://localhost:8001/index.html")

    print("✅ Navigation started! Open http://localhost:8001/index.html")

else:
    print("❌ Could not find valid coordinates for the locations.")
```
